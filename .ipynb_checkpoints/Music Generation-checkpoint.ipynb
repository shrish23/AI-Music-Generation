# Preprocessing all Files

```python
notes = []

for file in glob.glob("midi_songs/*.mid"):
    midi = converter.parse(file) # Converts file into stram.Score object
    
    print("parsing %s"%file)
    
    elements_to_parse = midi.flat.notes
    
    for ele in elements_to_parse:
        # if the element is a note, thne store its pitch
        if isinstance(ele,note.Note):
            notes.append(str(ele.pitch))

        # if the element is a chord, split each note of chord and join them with +
        elif isinstance(ele,chord.Chord):
            notes.append("+".join(str(n) for n in ele.normalOrder))
```

```python
notes
```

```python
with open('notes.pkl','wb') as filepath:
    pickle.dump(notes,filepath)
```

```python
with open("notes.pkl",'rb') as f:
    notes = pickle.load(f)
```

```python
n_vocab = len(set(notes))
```

```python
print("Total notes- ",len(notes))
print("Unique notes- ",n_vocab)
```

```python
print(notes[:100])
```

## Prepare Sequential Data for LSTM

```python
# How many LSTM input should consider
sequence_length = 100
```

```python
pitchnames = sorted(set(notes))
```

```python
element_to_int = dict((ele,num) for num,ele in enumerate(pitchnames))
```

```python
network_input = []
network_output = []
```

```python
for i in range(len(notes) - sequence_length):
    seq_in = notes[i : i+sequence_length] #contains 100 values
    seq_out = notes[i+sequence_length]
    
    network_input.append([element_to_int[ch] for ch in seq_in])
    network_output.append([element_to_int[seq_out]])
```

```python
# no. of examples
n_patterns = len(network_input)
print(n_patterns)
```

```python
network_input = np.reshape(network_input,(n_patterns,sequence_length,1))
```

```python
print(network_input.shape)
```

```python
#normalize data
normalized_network_input = network_input/float(n_vocab)
```

```python
# Network output are classes encode into one hot vector
```

```python
network_output = np_utils.to_categorical(network_output)
```

```python
network_output.shape
```

```python
print(normalized_network_input.shape)
print(network_output.shape)
```

## Create Model

```python
from keras.models import Sequential,load_model
from keras.layers import *
from keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
model = Sequential()
model.add(LSTM(512,input_shape=(normalized_network_input.shape[1],normalized_network_input.shape[2]),return_sequences = True))
model.add(Dropout(0.3))
model.add(LSTM(512,return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512))
model.add(Dense(256))
model.add(Dropout(0.3))
model.add(Dense(n_vocab,activation='softmax'))
model.summary()
```

```python
model.compile(loss="categorical_crossentropy",optimizer='adam')
```

```python
checkpoint = ModelCheckpoint("model.hdf5",monitor='loss',verbose=0,save_best_only=True,mode='min')

model_hist = model.fit(normalized_network_input,network_output,epochs=100,batch_size=64,callbacks=[checkpoint])
```

```python
model = load_model("new_weights.hdf5")
```

## Predictions

```python
sequence_length = 100
network_input = []
for i in range(len(notes) - sequence_length):
    seq_in = notes[i : i+sequence_length] #contains 100 values    
    network_input.append([element_to_int[ch] for ch in seq_in])
```

```python
start = np.random.randint(len(network_input) - 1)

# Mapping int_to_ele
int_to_ele = dict((num,ele) for num,ele in enumerate(pitchnames))

#initial Pattern
pattern = network_input[start]
prediction_output = []

#generate 200 elements
for note_index in range(200):
    prediction_input = np.reshape(pattern,(1,len(pattern),1))
    prediction_input = prediction_input/float(n_vocab)
    
    prediction = model.predict(prediction_input,verbose=0)
    
    idx = np.argmax(prediction)
    result = int_to_ele[idx]
    prediction_output.append(result)
    
    pattern.append(idx)
    pattern = pattern[1:]
```

```python
prediction_output
```

## Create Midi File

```python
offset = 0 # Time
output_notes = []

for pattern in prediction_output:
    
    # if pattern is a Chord
    if('+' in  pattern) or pattern.isdigit():
        notes_in_chord = pattern.split('+')
        tmp_notes = []
        for curr_note in notes_in_chord:
            new_note = note.Note(int(curr_note)) #create Note object for each note in the chord
            new_note.storedInstrument = instrument.Piano()
            tmp_notes.append(new_note)
            
        new_chord = chord.Chord(tmp_notes) #creates the chord from the list of notes
        new_chord.offset = offset
        output_notes.append(new_chord)
        
    else:
        #if pattern is a note
        new_note = note.Note(pattern)
        new_note.offset = offset
        new_note.storedInstrument = instrument.Piano()
        output_notes.append(new_note)
        
    offset += 0.5
```

```python
output_notes
```

```python
#create a stream object from the generated notes
midi_stream = stream.Stream(output_notes)
midi_stream.write('midi',fp="test_output.mid")
```
